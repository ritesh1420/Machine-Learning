# Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 1. prepare data
x=[3,4,5,6,7]
y=[6,8,10,12,14]

# 2. setup plot
fig,ax=plt.subplots(figsize=(10,8)) #(width,height)

# 3. plot data
ax.plot(x,y,'-*r')  # - for line,* for start ,r for red

# 4. Customize plot
ax.set(title="simple plot",
      xlabel="x-axis",ylabel="y-axis")

# 5. save & show(we save the the hole figure)

fig.savefig("image/sample-plot.png")
```

# Two options for subplots

```python
#subplot option 1
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,
                                       ncols=2,
                                       figsize=(10,5))
#plot to each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(food_price.keys(),food_price.values());
ax4.hist(np.random.randn(1000));
```

```python
#subplot option 1
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,
                                       ncols=2,
                                       figsize=(10,5))
#plot to each different axis
ax1.plot(x,x/2);
ax1.set(title="X vs X/2")

ax2.scatter(np.random.random(10),np.random.random(10));

ax3.bar(food_price.keys(),food_price.values());

ax4.hist(np.random.randn(1000));
```

```python
#Subplots option 2
fig,ax=plt.subplots(nrows=2,
                   ncols=2,
                   figsize=(10,5))
```

```python
#Subplots option 2
fig,ax=plt.subplots(nrows=2,
                   ncols=2,
                   figsize=(10,5))

#plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(food_price.keys(),food_price.values());
ax[1,1].hist(np.random.random(100));
```

# Which one we should use?(pyplot vs matplotlib oo method?)

1. When ploting something quickly,ok to use pyplot method
2. When plotting something more advance use the OO method

```python
heart_disease.head()
```

```python
over_50=heart_disease[heart_disease['age']>50]
over_50.head()
```

```python
len(over_50)
```

```python
#pyplot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            figsize=(10,5)); #c for colour
```

```python
#OO method
fig,ax=plt.subplots(figsize=(10,5))
over_50.plot( kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);
#ax.set_xlim([45,100]); for setting x limit
```
